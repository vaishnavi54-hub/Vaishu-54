FILE HANDLING:File handling is used to create, read, write, and update files in Python.
It allows Python programs to store data permanently.

Mode
Meaning
r
Read file
w
Write file (overwrites)
a
Append file
x
Create new file
r+
Read + Write

```python
# Write
with open("sample.txt","w") as f:
    f.write("Hello Vaishu\n")
    f.write("Welcome to Python\n")

# Read
with open("sample.txt","r") as f:
    print("Reading file:")
    print(f.read())

# Append
with open("sample.txt","a") as f:
    f.write("Appending new line\n")

# Read again
with open("sample.txt","r") as f:
    print("After append:")
    print(f.read())
```
