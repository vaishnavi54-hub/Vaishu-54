# USER DEFINED FUNCTIONS:A user-defined function is a function created by the programmer using the def keyword to perform a specific task.

```python
# 1. No arguments, no return value
def greet():
    print("Welcome to Python")

greet()


# 2. With arguments, no return
def greet_name(name):
    print("Hello", name)

greet_name("Vaishu")


# 3. No arguments, return value
def get_number():
    return 10

print("Returned value:", get_number())


# 4. With arguments and return value
def add(a, b):
    return a + b

print("Sum:", add(5, 3))


# 5. Default arguments
def student(name="Student"):
    print("Name:", name)

student()
student("Vaishu")


# 6. Keyword arguments
def details(name, age):
    print("Name:", name)
    print("Age:", age)

details(age=20, name="Vaishu")


# 7. Variable length arguments (*args)
def total(*nums):
    s = 0
    for i in nums:
        s += i
    return s

print("Total:", total(1, 2, 3, 4))


# 8. Variable length keyword arguments (**kwargs)
def info(**data):
    for key, value in data.items():
        print(key, ":", value)

info(course="B.Tech", year=3)


# 9. Recursive user-defined function
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print("Factorial:", factorial(5))


# 10. User-defined function with list
def max_element(lst):
    m = lst[0]
    for i in lst:
        if i > m:
            m = i
    return m

print("Maximum:", max_element([10, 25, 5, 40]))
```
