```python
import math          # import
from math import pi  # from, import

# global keyword
x = 10

def keyword_demo(a, b):
    global x          # global
    x = 20

    # if, elif, else
    if a > b and a != 0:     # if, and, !=
        print("a is greater")
    elif a == b or b == 0:  # elif, or
        print("Both are equal")
    else:
        print("b is greater")

    # for, break, continue
    for i in range(5):
        if i == 2:
            continue
        if i == 4:
            break
        print("i =", i)

    # while, pass
    j = 0
    while j < 2:
        pass
        j += 1

    # try, except, finally
    try:
        c = a / b
    except ZeroDivisionError:
        print("Division error")
    finally:
        print("Done division")

    # return
    return a + b


# class, def
class Demo:
    def show(self):
        print("Inside class")


# lambda
square = lambda n: n * n

# True, False, None
flag = True
value = None

# is, not, in
if flag is True and value is None:
    print("Keywords working")

# with, as
with open("demo.txt", "w") as f:
    f.write("Python keywords demo")

# assert
assert x == 20

# del
del flag

# raise
def check_positive(n):
    if n < 0:
        raise ValueError("Negative number")

# yield
def gen():
    yield 1
    yield 2

# nonlocal
def outer():
    y = 5
    def inner():
        nonlocal y
        y = 10
    inner()
    return y


# ---- Function Calls ----
print("Sum:", keyword_demo(10, 5))

obj = Demo()
obj.show()

print("Square:", square(4))

for val in gen():
    print(val)

print("Nonlocal value:", outer())
```
