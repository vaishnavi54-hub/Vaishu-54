BUILT IN MODULES:Built-in modules are pre-written Python files that contain functions and classes.
We import them to use ready-made features.

```python
import math
import random
import datetime
import sys
import os

print("Square root:", math.sqrt(36))
print("Factorial:", math.factorial(5))

print("Random number:", random.randint(1,100))

print("Today date:", datetime.date.today())

print("Python version:", sys.version)

print("Current folder:", os.getcwd())
```
