## RETURN STATEMENT:The return statement is used to send a value back from a function to the place where the function was called.
It ends the function
It gives output from function
A function can return one or more values

```python
# 1. Return single value
def add(a, b):
    return a + b

# 2. Return multiple values
def details():
    name = "Vaishu"
    age = 19
    return name, age

# 3. Return with condition
def check_even(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

# 4. Return inside loop
def find_square(n):
    return n * n

# 5. Recursive return
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

# Calling functions
print("Addition:", add(5, 3))

n, a = details()
print("Name:", n)
print("Age:", a)

print("Number is:", check_even(7))

print("Square:", find_square(4))

print("Factorial:", factorial(5))
```
