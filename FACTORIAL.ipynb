```python
# 1. Factorial using recursion
def factorial(n):
    if n == 1 or n == 0:
        return 1
    return n * factorial(n - 1)


# 2. Fibonacci series using recursion
def fibonacci(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)


# 3. Sum of first n natural numbers
def sum_n(n):
    if n == 0:
        return 0
    return n + sum_n(n - 1)


# 4. Power of a number (a^b)
def power(a, b):
    if b == 0:
        return 1
    return a * power(a, b - 1)


# 5. Count digits of a number
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)


# 6. Reverse a number using recursion
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    return reverse_number(n // 10, rev * 10 + n % 10)


# 7. Check palindrome using recursion
def palindrome(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return palindrome(s[1:-1])


# 8. Sum of digits
def sum_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_digits(n // 10)


# 9. Print numbers from 1 to n
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n, end=" ")


# 10. GCD of two numbers
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a % b)

print("Factorial of 5:", factorial(5))

print("Fibonacci series:")
for i in range(7):
    print(fibonacci(i), end=" ")

print("\nSum of first 10 numbers:", sum_n(10))

print("2 power 5:", power(2, 5))

print("Count digits in 12345:", count_digits(12345))

print("Reverse of 1234:", reverse_number(1234))

print("Is 'madam' palindrome?", palindrome("madam"))

print("Sum of digits of 123:", sum_digits(123))

print("Numbers from 1 to 5:")
print_numbers(5)

print("\nGCD of 48 and 18:", gcd(48, 18))
```
