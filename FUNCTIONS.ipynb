```python
# 1. Simple function
def greet():
    print("Hello, welcome to Python Functions")

greet()


# 2. Function with arguments
def greet_name(name):
    print("Hello", name)

greet_name("Vaishu")


# 3. Function with return value
def add(a, b):
    return a + b

print("Sum:", add(10, 20))


# 4. Multiple return values
def operations(a, b):
    return a+b, a-b, a*b, a/b

s, d, m, div = operations(10, 5)
print("Add:", s)
print("Sub:", d)
print("Mul:", m)
print("Div:", div)


# 5. Default argument
def college(name="Student"):
    print("Name:", name)

college()
college("Vaishu")


# 6. Keyword arguments
def details(name, age):
    print("Name:", name)
    print("Age:", age)

details(age=20, name="Vaishu")


# 7. Variable length arguments (*args)
def total(*nums):
    sum = 0
    for i in nums:
        sum += i
    return sum

print("Total:", total(1, 2, 3, 4, 5))


# 8. Variable length keyword arguments (**kwargs)
def info(**data):
    for key, value in data.items():
        print(key, ":", value)

info(name="Vaishu", course="B.Tech", year=3)


# 9. Lambda function
square = lambda x: x * x
print("Square:", square(6))


# 10. Recursive function
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print("Factorial:", factorial(5))


# 11. Function inside function
def outer():
    print("Outer function")
    def inner():
        print("Inner function")
    inner()

outer()


# 12. Function using list
def max_element(lst):
    return max(lst)

numbers = [10, 45, 23, 89, 12]
print("Maximum:", max_element(numbers))


# 13. Even or Odd using function
def even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print("7 is", even_odd(7))


# 14. Prime number function
def is_prime(n):
    if n <= 1:
        return "Not Prime"
    for i in range(2, n):
        if n % i == 0:
            return "Not Prime"
    return "Prime"

print("11 is", is_prime(11))


print("---- End of All-in-One Functions Program ----")
```
