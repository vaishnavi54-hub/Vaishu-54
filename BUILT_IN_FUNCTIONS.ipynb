# BUILT IN FUNCTIONS:Built-in functions are predefined functions provided by Python that can be used directly without importing any module.

```python
# Numeric functions
print("abs:", abs(-10))
print("pow:", pow(2, 3))
print("round:", round(3.456, 2))
print("max:", max(10, 20, 5))
print("min:", min(10, 20, 5))
print("sum:", sum([1, 2, 3, 4]))

# Type conversion functions
print("int:", int("10"))
print("float:", float(5))
print("str:", str(100))
print("bool:", bool(1))
print("list:", list((1, 2, 3)))
print("tuple:", tuple([4, 5, 6]))
print("set:", set([1, 2, 2, 3]))
print("dict:", dict(a=1, b=2))

# Sequence functions
print("len:", len("Python"))
print("range:", list(range(1, 6)))
print("sorted:", sorted([3, 1, 4, 2]))
print("reversed:", list(reversed([1, 2, 3])))
print("enumerate:", list(enumerate(['a', 'b', 'c'])))
print("zip:", list(zip([1, 2], ['x', 'y'])))

# Logical functions
print("all:", all([True, True, False]))
print("any:", any([False, False, True]))
print("isinstance:", isinstance(10, int))

# Character functions
print("chr:", chr(65))
print("ord:", ord('A'))

# Object & memory functions
x = 10
print("id:", id(x))
print("type:", type(x))
print("dir:", dir(x))

# Input / Output
# name = input("Enter name: ")
# print("Hello", name)
print("print function used")

# Evaluation functions
print("eval:", eval("5 + 5"))

# Attribute functions
class Demo:
    a = 10

obj = Demo()
print("hasattr:", hasattr(obj, 'a'))
print("getattr:", getattr(obj, 'a'))
setattr(obj, 'b', 20)
print("setattr:", obj.b)
delattr(obj, 'b')

# Callable
print("callable:", callable(print))

# Map and Filter
nums = [1, 2, 3, 4, 5]
print("map:", list(map(lambda x: x * 2, nums)))
print("filter:", list(filter(lambda x: x % 2 == 0, nums)))
```
