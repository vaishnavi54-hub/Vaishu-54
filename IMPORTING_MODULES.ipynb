IMPORTING MODULES:Importing a module means bringing functions and classes from another file/module into your program.

```python
# full module
import math
print("Square root:", math.sqrt(64))

# specific function
from random import randint
print("Random number:", randint(1,10))

# alias
import datetime as dt
print("Today:", dt.date.today())

# import all
from math import *
print("PI value:", pi)
```
