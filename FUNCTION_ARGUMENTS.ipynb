# FUNCTION ARGUMENTS:
Arguments are values passed to a function when calling it.
Types:
Positional arguments,
Keyword arguments,
Default arguments&
Variable length arguments

```python
# Positional
def add(a, b):
    print("Addition:", a+b)
add(5, 3)

# Keyword
def details(name, age):
    print(name, age)
details(age=19, name="Vaishu")

# Default
def greet(name="Student"):
    print("Hello", name)
greet()
greet("Vaishu")

# *args
def total(*nums):
    print("Sum:", sum(nums))
total(1,2,3,4)

# **kwargs
def student_info(**data):
    for k,v in data.items():
        print(k, ":", v)
student_info(name="Vaishu", age=19, dept="CSE")
```
