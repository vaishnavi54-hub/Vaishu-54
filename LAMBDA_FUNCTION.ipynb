LAMBDA FUNCTON:A lambda function is a small anonymous function (function without name).
It is used for short operations and written in one line.

```python
# 1. Simple lambda
add = lambda a, b: a + b
print("Addition:", add(5, 3))

# 2. Square
square = lambda x: x * x
print("Square:", square(4))

# 3. Even or Odd
check = lambda n: "Even" if n % 2 == 0 else "Odd"
print("7 is:", check(7))

# 4. Using lambda with map()
nums = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x*x, nums))
print("Squares list:", squares)

# 5. Using lambda with filter()
even_nums = list(filter(lambda x: x % 2 == 0, nums))
print("Even numbers:", even_nums)

# 6. Using lambda with sorted()
students = [("Ram", 50), ("Sam", 80), ("Asha",
```
