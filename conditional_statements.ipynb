# if:Runs code only if the condition is true.

```python
age = 18

if age >= 18:
    print("You are an adult")
```

# ifelse:chooses between two paths.

```python
age = 16

if age >= 18:
    print("You are an adult")
else:
    print("You are a minor")
```

# if elif else :Checks multiple conditions in order.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 70:
    print("Grade B")
elif marks >= 50:
    print("Grade C")
else:
    print("Fail")
```

## Nested if statements :An if inside another if

```python
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Entry allowed")
    else:
        print("ID required")
else:
    print("Too young")
```
